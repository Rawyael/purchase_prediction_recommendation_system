# Imports:


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data = pd.read_csv('E-commerece_sales_data_2024.csv')
sales_data
```

# Preprocessing (sales_data):

## droping duplicates:

```python
sales_data.columns
```

```python
sales_data.duplicated()
```

```python
sales_data.duplicated().sum()
```

```python
sales_data = sales_data.drop_duplicates()
sales_data
```

## missing data:

```python
sales_data = sales_data.drop(columns = ['Unnamed: 4'])
sales_data
```

```python
sales_data.isnull().sum().sort_values(ascending=False) 
```

```python
sales_data['Interaction type'] = sales_data['Interaction type'].fillna(sales_data['Interaction type'].mode()[0])
sales_data['Interaction type'] 
```

```python
sales_data['Interaction type'] = sales_data['Interaction type'].fillna(sales_data['Interaction type'].mode()[0])
```

```python
sales_data['Time stamp'] = sales_data['Time stamp'].fillna(sales_data['Time stamp'].mode()[0])
sales_data['Time stamp']
```

```python
sales_data['product id'] = sales_data['product id'].fillna(sales_data['product id'].mode()[0])
sales_data['product id']
```

```python
sales_data['user id'] = sales_data['user id'].fillna(sales_data['user id'].mode()[0]).round().astype(int)
sales_data['user id']
```

```python
sales_data[['Date', 'Time']] = sales_data['Time stamp'].str.split(' ', expand=True)
sales_data[['Date', 'Time']]
```

```python
sales_data
```

```python
sales_data = sales_data.drop(columns = ['Time stamp'])
sales_data
```

## visualizing :

```python
sns.countplot(data=sales_data, x='Interaction type')
```

```python
class_counts = sales_data['Interaction type'].value_counts()

# Print class counts
print(class_counts)

plt.bar(class_counts.index, class_counts.values)
```

# preprocessing (product_data):

```python
product_data = pd.read_csv('product_details.csv')
product_data
```

 ## droping duplicates:

```python
product_data.duplicated()
```

```python
product_data.duplicated().sum()
```

## missing data:

```python
product_data.isnull().sum().sort_values(ascending=False) 
```

```python
columns_to_drop = ['Product Description', 'Sku', 'Brand Name', 'Asin', 'Size Quantity Variant',
                   'List Price', 'Direction To Use', 'Quantity', 'Ingredients', 'Color',
                   'Dimensions', 'Product Details', 'Stock', 'Upc Ean Code', 'Product Dimensions',
                   'Variants', 'Model Number', 'Product Specification', 'Shipping Weight']
product_data = product_data.drop(columns=columns_to_drop)
```

```python
product_data
```

```python
columns_to_fill = ['Uniqe Id', 'Product Name', 'Category', 'Selling Price', 'About Product', 
                   'Technical Details', 'Image', 'Product Url', 'Is Amazon Seller']

# Fill missing values in specified columns with mode
product_data[columns_to_fill] = product_data[columns_to_fill].fillna(product_data[columns_to_fill].mode().iloc[0])
```

```python
product_data
```

```python
product_data.isnull().sum().sort_values(ascending=False) 
```

```python
product_data['Selling Price'].dtype
```
