```python
def clean_text(text):
    # Remove strange characters except for decimal point
    text = re.sub(r'[^a-zA-Z0-9\s\-.]', '', text)
    # Remove extra spaces and dashes
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'-+', '-', text)
    # Remove leading and trailing spaces
    text = text.strip()
    return text

columns_to_clean = ['Product_Description', 'Gender', 'Location', 'Product Name', 'Interaction type', 'Selling Price']

for column in columns_to_clean:
    df1[column] = df1[column].apply(clean_text)
```

```python
# Step 1: Remove strange characters and spaces
def clean_price(price):
    # Check if the value is NaN
    if pd.isna(price):
        return np.nan
    # Remove non-numeric characters and spaces
    price = re.sub(r'[^0-9.]', '', str(price))
    return price

# Clean 'Selling Price' column
df2['Selling Price'] = df2['Selling Price'].apply(clean_price)

# Step 2: Convert to numeric
df2['Selling Price'] = pd.to_numeric(df2['Selling Price'], errors='coerce')  # 'coerce' to handle any conversion errors
```

```python
import numpy as np

# Calculate quartiles
Q1 = np.percentile(df2['Selling Price'], 25)
Q3 = np.percentile(df2['Selling Price'], 75)

# Calculate IQR
IQR = Q3 - Q1

# Define threshold for outliers
threshold = 1.5 * IQR

# Identify outliers
outliers = df2['Selling Price'][(df2['Selling Price'] < Q1 - threshold) | (df2['Selling Price'] > Q3 + threshold)]

print("Outliers:", outliers)
```

```python
from sklearn.preprocessing import LabelEncoder

# Splitting data into features and target
X = df2.drop(columns='Interaction type')
y = df2['Interaction type']

# Label encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

# Defining preprocessing for numerical and categorical features
numeric_features = ['Age', 'Selling Price']
categorical_features = ['Gender', 'Location', 'Product Name']
text_feature = 'Product_Description'  # Note this is now a single string, not a list

# Preprocessing for numerical features (Age and Selling Price)
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Preprocessing for categorical features (Gender, Location, Category_y, Product Name)
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Define the ColumnTransformer without the text transformation
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Extract text feature for TF-IDF transformation
text_data_train = X_train[text_feature]
text_data_test = X_test[text_feature]  # Fix here, use X_test for text data

# Apply TF-IDF Vectorizer to text data
tfidf_transformer = TfidfVectorizer()
text_data_train_tfidf = tfidf_transformer.fit_transform(text_data_train)
text_data_test_tfidf = tfidf_transformer.transform(text_data_test)

# Preprocess the non-text features
X_train_preprocessed = preprocessor.fit_transform(X_train.drop(columns=text_feature))
X_test_preprocessed = preprocessor.transform(X_test.drop(columns=text_feature))

# Concatenate the TF-IDF vectorized text feature with the other preprocessed features
X_train_final = hstack([X_train_preprocessed, text_data_train_tfidf])
X_test_final = hstack([X_test_preprocessed, text_data_test_tfidf])
```
