# product table

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
product_data = pd.read_csv('product_details.csv')
product_data
```

# preprocessing

```python
product_data.duplicated()
```

```python
product_data.isnull().sum().sort_values(ascending=False) 
```

```python
columns_to_drop = ['Product Description', 'Sku', 'Brand Name', 'Asin', 'Size Quantity Variant',
                   'List Price', 'Direction To Use', 'Quantity', 'Ingredients', 'Color',
                   'Dimensions', 'Product Details', 'Stock', 'Upc Ean Code', 'Product Dimensions',
                   'Variants', 'Model Number', 'Product Specification', 'Shipping Weight']
product_data = product_data.drop(columns=columns_to_drop)
```

```python
product_data
```

```python
columns_to_fill = ['Uniqe Id', 'Product Name', 'Category', 'Selling Price', 'About Product', 
                   'Technical Details', 'Image', 'Product Url', 'Is Amazon Seller']

# Fill missing values in specified columns with mode
product_data[columns_to_fill] = product_data[columns_to_fill].fillna(product_data[columns_to_fill].mode().iloc[0])
```

```python
product_data
```

```python
product_data.isnull().sum().sort_values(ascending=False)
```

# Creating the TF-IDF Matrix

```python
product_data['Product Name']
```

```python
product_data['Product_Description'] = product_data['About Product'] + ' ' + product_data['Technical Details'] + ' ' + product_data['Category']
```

```python
product_data['Product_Description']
```

```python
product_data
```

```python
product_data.drop(columns=['About Product', 'Technical Details', 'Category'], inplace=True)
```

```python
product_data
```

```python
product_data["Product_Description"] = product_data["Product_Description"].str.replace(r"[^\w\s]", " ", regex=True).str.replace(r"[\d]", " ", regex=True)
```

```python
product_data["Product_Description"]
```

```python
tfidf = TfidfVectorizer(stop_words="english", min_df=4)

tfidf_matrix = tfidf.fit_transform(product_data["Product_Description"])
```

```python
similarity = cosine_similarity(tfidf_matrix,tfidf_matrix)
```
